### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

try:
    # Bar Plot with Colors
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['red', 'blue', 'green', 'purple']

    # Create the bar plot
    plt.figure(figsize=(8, 5))
    plt.bar(categories, values, color=colors)

    # Customize the bar plot
    plt.title('Bar Plot with Different Colors')
    plt.xlabel('Categories')
    plt.ylabel('Values')

    # Show the bar plot
    plt.show()

except ImportError as e:
    print(f"ImportError: {e}. Please make sure you have the required libraries installed.")
except ValueError as e:
    print(f"ValueError: {e}. Check the data passed to the bar plot.")
except Exception as e:
    print(f"An error occurred in the bar plot: {e}")
```
