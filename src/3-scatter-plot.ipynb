### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Scatter Plot with Annotations
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, color='red')

# Annotate each point
for i in range(len(x_scatter)):
    plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customize scatter plot
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the scatter plot
plt.show()
```
