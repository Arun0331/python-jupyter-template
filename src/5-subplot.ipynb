### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data for the charts
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
categories = ['A', 'B', 'C', 'D', 'E']
sizes = [25, 35, 15, 10, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line chart in the first subplot
axs[0, 0].plot(x, y, marker='o', color='b')
axs[0, 0].set_title('Line Chart')

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, sizes, color='g')
axs[0, 1].set_title('Bar Chart')

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x, y, color='r')
axs[1, 0].set_title('Scatter Plot')

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
